```python
# Loading dataset two.
df2 = pd.read_csv('bse_dataset2.csv')
df2
```

```python
# Check null values in dataset two
df2.isnull().sum().sort_values(ascending = False)
```

```python
# Checking for basic information in the second dataset
df2.info()
```

```python
# Checking null values in percentages
for i in df2.columns:
    if df2[i].isnull().any():
        print(i, ":", df2[i].isnull().sum()*100/df2.shape[0], "%")
```

```python
# Checking null values in figures
import missingno as msno
msno.bar(df2, color = "gray", fontsize = 12, figsize = (10, 5))
```
