# DATA ANALYSIS WITH PANDAS.

**A. SERIES.**

**QN.1  Comparing elements of two Pandas series.**

```python
import pandas as pd
import numpy as np

# Creating the first Pandas Series
ser = pd.Series([4, 65, 436, 3, 9])

# Creating the second Pandas Series
seri = pd.Series([7, 0, 3, 897, 9])

# Comparing the elements of the two series
comparison = ser == seri

print(comparison)
```

**QN.2  Computations on series.**

```python
s1 = pd.Series([2, 4, 6, 8, 14])
s2 = pd.Series([1, 5, 5, 7, 9])

# Adding the two series
addition = s1 + s2

# Subtracting the two series
subtraction = s1 - s2

# Multiplying the two series
multiplication = s1 * s2

# Dividing the two series
division = s1 / s2

# Displaying the results
print("Addition:")
print(addition)
print("\nSubtraction:")
print(subtraction)
print("\nMultiplication:")
print(multiplication)
print("\nDivision:")
print(division)
```

**QN.3  Converting a dictionary to a Pandas Series.**

```python
my_d = {'Josh': 24, 'Sam': 36, 'Peace': 19, 'Charles': 65, 'Tom': 44}

s = pd.Series(my_d)

# Displaying the resulting Pandas Series
s
```

**QN.4 Converting a series to an array.**

```python
series1 = pd.Series(['Love', 800, 'Joy', 789.9, 'Peace', True])

array1 = np.array(series1)

# Displaying the resulting array
print(array1)
```

**QN.5 Display the most frequent value in the given series and replace everything else as ‘Other’.**

```python
# Since every column in a dataframe is a Series, the series is the HomeTeamGoals.
df = pd.read_csv('AfricaCupofNationsMatches.csv')

# I want to know which unique values are present in the series. 
df['HomeTeamGoals'].unique()
```

```python
# Find the most frequent value.
most_frequent = df['HomeTeamGoals'].mode()[0]
most_frequent
```

```python
# Replacing all values in the Series that are not the mode with 'Other'
df['HomeTeamGoals'].apply(lambda x: 'Other' if x != most_frequent else x)
```
