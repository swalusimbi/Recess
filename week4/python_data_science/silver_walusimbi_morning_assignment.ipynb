**Computational Logic to deal with NaN in the dataset**

```python
# Generating grades array [1,2,3,4,7,9,X] from each percentage grade
grades = []
for column in df.columns:
    if (column.startswith("%")):
        grades.append(column[-1])

# Generating coefficient of key sections.
for row in df.index:
    total_coefficient = df.at[row, "TOTAL_CANDIDATES"] / 100
    male_coefficient = df.at[row, "MALE_CANDIDATES"] / 100
    female_coefficient = df.at[row, "FEMALE_CANDIDATES"] / 100

    # For each grade [1,2,3,4,7,9,X] obtain its total, %, male%, female%, maleTotal and femaleTotal
    for grade in grades:
        total = list(filter(lambda x:x.endswith(grade) and x.startswith("TOTAL"),df.columns))[0]
        totalPercentage = list(filter(lambda x:x.endswith(grade) and x.startswith("%"),df.columns))[0]
        maleTotal = list(filter(lambda x:x.endswith(grade) and x.startswith("MALE") and "TOTAL" in x,df.columns))[0]
        femaleTotal = list(filter(lambda x:x.endswith(grade) and x.startswith("FEMALE") and "TOTAL" in x,df.columns))[0]
        malePercentage = list(filter(lambda x:x.endswith(grade) and x.startswith("MALE") and "%" in x,df.columns))[0]
        femalePercentage = list(filter(lambda x:x.endswith(grade) and x.startswith("FEMALE") and "%" in x,df.columns))[0]

        # Whenever the coefficient is 0, fill the remaining related fields for that grade with 0's.
        if(total_coefficient == 0.0):
            df.at[row,total] = 0.0
            df.at[row,totalPercentage] = 0.0
        if(male_coefficient == 0.0):
            df.at[row,maleTotal] = 0.0
            df.at[row,malePercentage] = 0.0
        if(female_coefficient == 0.0):
            df.at[row,femaleTotal] = 0.0
            df.at[row,femalePercentage] = 0.0

        # Whenever percentage grade is 0.0%, replace its corresponding NaN grade with 0.0
        if(df.at[row, totalPercentage] == 0.0 and pd.isna(df.at[row, total])):
            df.at[row,total] = 0.0

        if(df.at[row, femalePercentage] == 0.0 and pd.isna(df.at[row, femaleTotal])):
            df.at[row, femaleTotal] = 0.0

        if(df.at[row, malePercentage] == 0.0 and pd.isna(df.at[row, maleTotal])):
            df.at[row, maleTotal] = 0.0

        # Whenever grade 0.0 replace its corresponding NaN %grade with 0.0
        if(df.at[row, total] == 0.0 and pd.isna(df.at[row, totalPercentage])):
            df.at[row,totalPercentage] = 0.0

        if(df.at[row, femalePercentage] == 0.0 and pd.isna(df.at[row, femalePercentage])):
            df.at[row, femalePercentage] = 0.0

        if(df.at[row, malePercentage] == 0.0 and pd.isna(df.at[row, malePercentage])):
            df.at[row, malePercentage] = 0.0

        # Using the calculated coefficients.
        if(pd.notna(df.at[row, total]) and pd.isna(df.at[row, totalPercentage])):
            df.at[row, totalPercentage] = round((df.at[row,total] / total_coefficient), 1)

        if(pd.notna(df.at[row, totalPercentage]) and pd.isna(df.at[row, total])):
            df.at[row, total] = int(total_coefficient * df.at[row, totalPercentage])

        if(pd.notna(df.at[row, maleTotal]) and pd.isna(df.at[row, malePercentage])):
            df.at[row, malePercentage] = round((df.at[row,maleTotal] / male_coefficient), 1)

        if(pd.notna(df.at[row, malePercentage]) and pd.isna(df.at[row, maleTotal])):
            df.at[row, maleTotal] = int(male_coefficient * df.at[row, malePercentage])

        if(pd.notna(df.at[row, total]) and pd.isna(df.at[row, totalPercentage])):
            df.at[row, totalPercentage] = round((df.at[row,total] / total_coefficient), 1)

        if(pd.notna(df.at[row, femalePercentage]) and pd.isna(df.at[row, femaleTotal])):
            df.at[row, femaleTotal] = int(female_coefficient * df.at[row, femalePercentage])

        # Applying subtraction from Total %ages, to fill the missing ones i.e male and female percentages
        if(pd.notna(df.at[row, malePercentage]) or pd.notna(df.at[row, femalePercentage])) is not (pd.notna(df.at[row, femalePercentage]) == pd.notna(df.at[row, malePercentage])):
            if(pd.notna(df.at[row, malePercentage])):
                df.at[row, femaleTotal] = df.at[row, totalPercentage] - df.at[row, malePercentage]

            if(pd.notna(df.at[row, femalePercentage])):
                df.at[row, maleTotal] = df.at[row, totalPercentage] - df.at[row, femalePercentage]

        # Get missing total grades by summation and subtraction of male, female and general totals.
        if not (pd.notna(df.at[row, total]) and pd.notna(df.at[row,maleTotal]) and pd.notna(df.at[row,femaleTotal])):
            if(pd.notna(df.at[row,total]) and pd.notna(df.at[row,maleTotal])):
                df.at[row,femaleTotal] = df.at[row,total] - df.at[row,maleTotal]
            if(pd.notna(df.at[row,total]) and pd.notna(df.at[row,femaleTotal])):
                df.at[row,maleTotal] = df.at[row,total] - df.at[row,femaleTotal]
            if(pd.notna(df.at[row,maleTotal]) and pd.notna(df.at[row,femaleTotal])):
                df.at[row, total] = df.at[row,maleTotal] + df.at[row,femaleTotal]
```

**DATA TRAINING AND PREDICTION.**

DETERMINING FEATURES WITH HIGH CORRELATION ON TARGETS.

```python
# From all the columns.
# Choose the proposed features.
X = df.loc[:, ['YEAR','DISTRICT','SCHOOL','TOTAL_CANDIDATES','FEMALE_CANDIDATES','MALE_CANDIDATES']]
X
```

```python
# Encoding strings to Numerical values.
from sklearn.preprocessing import LabelEncoder
district_le = LabelEncoder()
X.loc[:, 'DISTRICT'] = district_le.fit_transform(X.DISTRICT)

school_le = LabelEncoder()
X.loc[:, 'SCHOOL'] = school_le.fit_transform(X.SCHOOL)

X
```

```python
# Checking datatypes of the features.
X.info()
```

```python
y = df[['FEMALE_%_DIV1','FEMALE_%_DIV2','FEMALE_%_DIV3','FEMALE_%_DIV4','FEMALE_%_DIV7','FEMALE_%_DIV9','FEMALE_%_X',
            'MALE_%_DIV1','MALE_%_DIV2','MALE_%_DIV3','MALE_%_DIV4','MALE_%_DIV7','MALE_%_DIV9','MALE_%_X']]

y 
```

**FEATURE SELECTION.**

```python
from sklearn.linear_model import MultiTaskLassoCV

reg = MultiTaskLassoCV()
reg.fit(X,y)
print('Best alpha using built-in MultiTaskLassoCV: %f' % reg.alpha_)
print('Best score using built-in MultiTaskLassoCV: %f' % reg.score(X,y)) 
```

```python
# Taking any random row from the coefficients array:  row (coefficient threshold: 7)
coef1 = pd.Series(reg.coef_[7], index=X.columns)

# Set the figure size
plt.rcParams['figure.figsize'] = (6.0, 8.0)

# Plot the coefficients as a horizontal bar chart
coef1.plot(kind='barh')
plt.title("Feature Importance using MultiTaskLasso Model")
plt.show()
```

```python
# As observed, year and total candidates are not suitable features for Modeling, thus eliminated since they get a score of 0

# Refined Dataset.
X.drop(['YEAR','TOTAL_CANDIDATES'], axis=1,inplace=True)
X 
```

**Create a MultiRegression Ridge Model.**

```python
from sklearn.multioutput import MultiOutputRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.1)

gbr = GradientBoostingRegressor()

# Create multiple parallel instances of the estimator.
model = MultiOutputRegressor(estimator=gbr).fit(X_train,y_train)

# Determine score of the model in approximating targets
score = model.score(X_train,y_train)
print('Training score: ', score)
```

```python
# Predictions of the targets
y_pred = model.predict(X_test)

# target column names
ylabel = y_test.columns

# Get MSE
for i in range(y_pred.shape[1]):
    print(ylabel[i]+' MSE:  %.4f\n' %mean_squared_error(y_pred[:,i],y_test[ylabel[i]]))
```

```python
# Relation between predicted and actual percentage grades.
for i in range(y_pred.shape[1]):
    fig, ax= plt.subplots()
    ax.scatter(y_test[ylabel[i]], y_pred[:,i])
    ax.plot([y_test[ylabel[i]].min(),y_test[ylabel[i]].max()],[y_test[ylabel[i]].min(),y_test[ylabel[i]].max()],'k--',lw=4)
    ax.set_xlabel("Actual %s" %ylabel[i])
    ax.set_ylabel("Predicted %s" %ylabel[i])
    plt.show()
```
