```python
# Filling missing values with mean, median and mode.
# Mean.
df = pd.read_csv('bse_dataset4.csv')
m = df['Age'].mean()
print(m)
df['Age'].fillna(m, inplace=True)
```

```python
# Return rows with the mean.
fil_df = df[df['Age'] == m]
fil_df
```

```python
# Median.
df = pd.read_csv('bse_dataset4.csv')
me = df['Age'].median()
print(me)
df['Age'].fillna(me, inplace=True)
```

```python
# Return rows with the median.
fill_df = df[df['Age'] == me]
fill_df
```

EXERCISE.1  Filling missing values with mode.

```python
# Mode.     iloc[0] is used to access the first row of the mode DataFrame.
df = pd.read_csv('bse_dataset4.csv')
md = df['Age'].mode().iloc[0]
print(md)
df['Age'].fillna(md, inplace=True)
```

```python
# Return rows with the mode.
fi_df = df[df['Age'] == md]
fi_df
```

```python
# Wrong format.
# Row at position 5 has NaN, so we fill it with a default date.
df['Timestamp'].fillna("05/07/2023 09:45:13", inplace=True)
df.iloc[5]
```

```python
# Convert the 'Timestamp' column to datetime format
df['Timestamp'] = pd.to_datetime(df['Timestamp'], errors='coerce')
df.head(15)
```

```python
# Wrong data.  .. 23000 instead of 23   ..  21.99 instead of 2199
for x in df.index:
    if df.loc[x, 'gender'] == 'Marabou Stock':
        df.drop(x, inplace=True)
```

```python
# Check if there are any duplicates
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
# Confirm that the duplicates are dropped
df.duplicated().sum()
```

```python
# Data Visualization. 
# Plotting heatmaps, piecharts, barplots, lineplots, scatterplots, histograms, boxplots.
```

```python
plt.hist(df['Age'], bins=30)
plt.xlabel('Age')
plt.ylabel('Number of students')
plt.title('Histogram')
plt.show()
```

```python
# Count the frequency of each age
age_counts = df['Age'].value_counts()

# Create the pie chart
plt.pie(age_counts, labels=age_counts.index, autopct='%1.1f%%')
plt.axis('equal')
plt.title('Age Distribution of Students.')
plt.show()
```

```python
# Count the frequency of each age
age_counts = df['Age'].value_counts().sort_index()

# Create the bar plot
plt.bar(age_counts.index, age_counts.values)
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Bar plot showing age distribution of students.')
plt.show()
```
